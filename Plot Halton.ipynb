```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data00 = np.genfromtxt("build/faure0_0.csv", delimiter=',')
data01 = np.genfromtxt("build/faure0_1.csv", delimiter=',')
data10 = np.genfromtxt("build/faure1_0.csv", delimiter=',')
data11 = np.genfromtxt("build/faure1_1.csv", delimiter=',')
```

```python
plt.figure(figsize=[10,10])
plt.subplot(2, 2, 1)

# 00
plt.scatter(data00[:,3], data00[:,4])
plt.title('Pixel[0,0]: dim(4,5)')

# 01
plt.subplot(2, 2, 2)
plt.scatter(data01[:,3], data01[:,4])
plt.title("Pixel[0,1]: dim(4, 5)")

# 10
plt.subplot(2, 2, 3)
plt.scatter(data10[:,3], data10[:,4])
plt.title("Pixel[1,0]: dim(4, 5)")

# 11
plt.subplot(2, 2, 4)
plt.scatter(data11[:,3], data11[:,4])
plt.title("Pixel[1,1]: dim(4, 5)")

plt.show()
```
