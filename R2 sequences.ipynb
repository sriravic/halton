```python
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
# define some constants for our use
# some constants for 3 dimensions
phi_1 = 1.6180339887498948482
phi_2 = 1.32471795724474602596
phi_3 = 1.22074408460575947536

# custom fmod(val, 1)
def mymod(val):
    return val - math.floor(val)
```

```python
def r2_2dim(n, shift = 0):
    a1 = 1.0 / phi_2
    a2 = 1.0 / (phi_2 * phi_2)
    x = mymod(0.5 + a1 * (n + shift))
    y = mymod(0.5 + a2 * (n + shift))
    return (x, y)
```

```python
# jittered r2 sequences
import math

delta0 = 0.76
i0 = 0.700
kpi = 3.142
rootpi = math.sqrt(kpi)

def r2_2dim_jittered(n, shift, lamb, beta0, beta1):
    # compute the original point
    pt = r2_2dim(n, shift)
    
    # compute the additional offset
    factor = lamb * delta0 * rootpi * 0.25 / (math.sqrt(n - i0))
    
    x = mymod(pt[0] + factor * beta0)
    y = mymod(pt[1] + factor * beta1)
    
    return (x, y)
```

```python
import random
npts = 10000
pts_r2_x = []
pts_r2_y = []
shift = 0
for n in range(1, npts):
    res = r2_2dim(n, shift)
    pts_r2_x.append(res[0])
    pts_r2_y.append(res[1])
```

```python
pts_r2_jx = []
pts_r2_jy = []
shift = 0
for n in range(1, npts):
    u0 = random.uniform(0.0, 1.0)
    u1 = random.uniform(0.0, 1.0)
    lamb = random.uniform(0.0, 1.0)
    res = r2_2dim_jittered(n, shift, lamb, u0, u1)
    pts_r2_jx.append(res[0])
    pts_r2_jy.append(res[1])
```

```python
markersize = 1.0
plt.figure(figsize=(20, 10))
plt.subplot(1, 2, 1)
plt.title('R2');
plt.scatter(pts_r2_x, pts_r2_y, s = markersize)

plt.subplot(1, 2, 2)
plt.title('R2-jittered')
plt.scatter(pts_r2_jx, pts_r2_jy, s = markersize)
plt.show()
```

```python
# map a 2d random number to a point on the sphere
def map2sphere(pt):
    z = 1.0 - 2.0 * pt[0]
    r = math.sqrt(max(0, 1.0 - z * z));
    phi = 2.0 * kphi * pt[1]
    return (r * math.cos(phi), r * math.sin(phi), z)

def map2hemisphere(pt):
    z = pt[0]
    r = math.sqrt(max(0, 1.0 - z * z));
    phi = 2.0 * kphi * pt[1]
    return (r * math.cos(phi), r * math.sin(phi), z)
    
```
